# Multivariable Linear Regression

```python
import numpy as np
import matplotlib.pyplot as mpl
```

# 1st Degree Polynomial Linear Regression

```python
## Read data from the text file
with open('rawdata.dat') as f:
    x = []
    y = []
    for line in f:
        temp = line.split()
        x.append([1,float(temp[0])])
        y.append(float(temp[1]))
    designMatrix = np.array(x)
    yMatrix = np.transpose(np.array(y))
```

```python
## solve for the coefficient matrix (B) = (xTx)^-1 * (xTy), then find the predicted values of Y = XB
def regress(designMatrix,y):
    xTranspose = np.transpose(designMatrix)
    xTx = np.matmul(xTranspose, designMatrix)
    xTxInv = np.linalg.inv(xTx)
    xTy = np.matmul(xTranspose, yMatrix)
    coefMatrix = np.matmul(xTxInv,xTy)
    return coefMatrix
```

```python
coefMatrix = regress(designMatrix,yMatrix)
yHat = np.matmul(designMatrix,coefMatrix)
```

```python
mpl.plot(np.transpose(designMatrix)[:][1],yHat)
mpl.scatter(np.transpose(designMatrix)[:][1],yMatrix)
mpl.title('Degree 1 Polynomial Linear Regression')
mSE = sum(((yMatrix-yHat)*(yMatrix-yHat))/yMatrix.size)
print('Degree 1 Polynomial Linear Regression')
print('Prediction = ' + str(coefMatrix[0])[0:6]+' + '+str(coefMatrix[1])[0:6] + 'x')
print("Mean Square Error: " + str(mSE))
print("Mean Error: "+ str(np.sqrt(mSE)))
```

# 2nd Degree Polynomial Linear Regression

```python
## Read data from the text file and convert it into the design matrix and the y value vector.
with open('rawdata.dat') as f:
    x = []
    y = []
    for line in f:
        temp = line.split()
        x2 = float(temp[0])*float(temp[0])
        x.append([1,float(temp[0]),x2])
        y.append(float(temp[1]))
    designMatrix = np.array(x)
    yMatrix = np.transpose(np.array(y))
```

```python
coefMatrix = regress(designMatrix,yMatrix)
yHat = np.matmul(designMatrix,coefMatrix)
```

```python
mpl.plot(np.transpose(designMatrix)[:][1],yHat)
mpl.scatter(np.transpose(designMatrix)[:][1],yMatrix)
mpl.title('Degree 2 Polynomial Linear Regression')
mSE = sum(((yMatrix-yHat)*(yMatrix-yHat))/yMatrix.size)
print('Degree 2 Polynomial Linear Regression')
print('Prediction = ' + str(coefMatrix[0])[0:6]+' + '+str(coefMatrix[1])[0:6] + 'x + ' + str(coefMatrix[2])[0:4] + 'x^2')
print("Mean Square Error: " + str(mSE))
print("Mean Error: "+ str(np.sqrt(mSE)))
```
